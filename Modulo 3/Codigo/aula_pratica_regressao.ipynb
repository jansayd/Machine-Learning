```python
import matplotlib.pyplot as plt 
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, median_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# carregando o dataset
x, y = datasets.load_diabetes(return_X_y = True)
```

```python
# separamos uma feature para usar
x = x[:, np.newaxis, 2]
```

```python
# separada os dados para treino e teste
x_train, x_teste, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
# instancia do modelo
regr = linear_model.LinearRegression()
```

```python
# treinando o modelo
regr.fit(x_train, y_train)
```

```python
# testando o modelo
y_pred = regr.predict(x_teste)
```

```python
# metricas de qualidade
print('Erro médio quadratico: %.2f' % mean_squared_error(y_test, y_pred))
print('Erro mediano absoluto: %.2f' % median_absolute_error(y_test, y_pred))
```

```python
# plot do grafico para representar o resultado
plt.scatter(x_teste, y_test, color = 'black')
plt.plot(x_teste, y_pred, color = 'blue', linewidth = 3)
plt.xticks()
plt.yticks()
plt.show()
```
